```python
from langchain_ollama import ChatOllama

llm = ChatOllama(
    model="llama3.2",
    num_predict=1024
    format="json"
)
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "You are a helpful assistant that translates {input_language} to {output_language}.",
        ),
        ("human", "{input}"),
    ]
)

chain = prompt | llm
chain.invoke(
    {
        "input_language": "English",
        "output_language": "Tamil",
        "input": "I love programming.",
    }
)
```

```python
from langchain_core.output_parsers import StrOutputParser

output_parser = StrOutputParser()
chain = chain | output_parser

chain.invoke({
        "input_language": "English",
        "output_language": "Tamil",
        "input": "The sky is blue",
    })
```

```python
st = chain.stream({
        "input_language": "English",
        "output_language": "Tamil",
        "input": "The sky is blue",
    })

for delta in st:
    print(delta, end="")
```

```python
from langchain_community.vectorstores import Chroma
from langchain_text_splitters import MarkdownTextSplitter

text_splitter = MarkdownTextSplitter()
documents = text_splitter.split_documents(docs)
vector = Chroma.from_documents(documents, embeddings, collection_name="test")
```

```python
results = vector.similarity_search(
    "Explain lists in Python",
    k=2,
    # filter={"source": "tweet"},
)
for res in results:
    print(f"* {len(res.page_content)} [{res.metadata}]")
```

```python
from langchain.chains.combine_documents import create_stuff_documents_chain

prompt = ChatPromptTemplate.from_template("""Answer the following question based only on the provided context:

<context>
{context}
</context>

Question: {input}""")

document_chain = create_stuff_documents_chain(llm, prompt)
```

```python
from langchain.chains import create_retrieval_chain

retriever = vector.as_retriever()
retrieval_chain = create_retrieval_chain(retriever, document_chain)
```

```python
response = retrieval_chain.invoke({"input": "Explain the difference between Iterators and Generators. Explain with an example"})
print(response["answer"])
```

```python
print(response["context"][0].page_content)
```

```python
import json
from langchain_ollama import ChatOllama
from langchain_core.prompts import ChatPromptTemplate

llm = ChatOllama(model="llama3.2", num_predict=1024, format="json")
system_prompt = """Given an one line description of a product generate mock data for an e-commerce site.
Make up dimensions and metrics if not given by the user.

The response must be of the following JSON format:
<format>
{{
  "title": "title for the product",
  "content": "product details like description, features and dimesions with neat heading, text highlighting",
  "tags": [5 relevant tags for search]
}}
</format>
"""

prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            system_prompt,
        ),
        ("human", "{input}"),
    ]
)

chain = prompt | llm
response = chain.invoke(
    {
        "input": "a black and white shoe with a black and white stripe ",
    }
)

content_dict = json.loads(response.content)
print(json.dumps(content_dict, indent=3))
```
